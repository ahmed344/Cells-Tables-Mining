### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

### The whole table

```python
df = pd.read_csv('Cells_on_BSA_Experiments.csv')
print(df.shape)
df
```

### Experiment on glass

```python
df_glass = df[df['Incubation'].str.contains('glass')]
print(f"N = {df_glass.shape[0]}, n = {df_glass['No'].sum() + df_glass['Small'].sum() + df_glass['Big'].sum()}")
df_glass
```

### Drop the glass row

```python
df = df[df['Incubation'].str.contains('glass') == False]
df.shape
```

### Experiments on just BSA

```python
df_bsa = df[df['Discription'].str.contains("BSA-NTA")==False]
print(f"N = {df_bsa.shape[0]}, n = {df_bsa['No'].sum() + df_bsa['Small'].sum() + df_bsa['Big'].sum()}")
df_bsa
```

### Experiments on BSA-NTA

```python
df_nta = df[df['Discription'].str.contains("BSA-NTA")]
print(f"N = {df_nta.shape[0]}, n = {df_nta['No'].sum() + df_nta['Small'].sum() + df_nta['Big'].sum()}")
df_nta
```

### SHC

```python
df_nta_shc = df_nta[df_nta['Incubation'].str.contains("she") == False]

print(f"N = {df_nta_shc.shape[0]}, n = {df_nta_shc['No'].sum() + df_nta_shc['Small'].sum() + df_nta_shc['Big'].sum()}")
df_nta_shc
```

```python
df_nta_shc_p = df_nta_shc[df_nta_shc['Incubation'].str.contains("without")==False]
print(f"With Ecad N = {df_nta_shc_p.shape[0]}, n = {df_nta_shc_p['No'].sum() + df_nta_shc_p['Small'].sum() + df_nta_shc_p['Big'].sum()}")

df_nta_shc_n = df_nta_shc[df_nta_shc['Incubation'].str.contains("without")]
print(f"Without Ecad N = {df_nta_shc_n.shape[0]}, n = {df_nta_shc_n['No'].sum() + df_nta_shc_n['Small'].sum() + df_nta_shc_n['Big'].sum()}")
```
